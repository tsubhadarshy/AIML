```python
import numpy as np
import pandas as pd
```

```python
laliga = pd.read_csv("Laliga.csv")
# please keep the laliga.csv file in working directory
```

```python
laliga.head()
# reading the file head for an overview
```

```python
laliga.rename(columns=laliga.iloc[0],inplace=True)
# making first row as column header for better accessibility
laliga
```

```python
laliga.drop([0],inplace=True)
# dropping first row to remove any confusion later
laliga
```

```python
laliga.set_index=('Pos')
# setting first column as index for simplicity
```

```python
laliga.describe(include="all")
# for arithmetic decription of all columns involved
```

```python
laliga.dtypes
# to inspect data type of each column involved
```

```python
# 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (5 points) 
laliga.replace('-',0,inplace=True)
laliga
```

```python
# 2. Print all the teams which have started playing between 1930-1980. Use “Debut” column  (5 points)
laliga['DebutInt']=laliga['Debut'].str[:4].astype('int64')
# typecasted Debut column as int64 into a new column DebutInt to be able to perform arithmetic comparison
laliga_debut = laliga[laliga.DebutInt.isin([1930,1980])]
# made a new DF with the filtered teams
print("The teams which have debuted in the years 1930-1980 are:",laliga_debut['Team'].values)
# printed the list with only the values of filtered series to remove the dtypes in answer to be printed
```

```python
# 3. Print the list of teams which came Top 5 in terms of points (2.5 points)
laliga['PointsInt']=laliga['Points'].astype('int64')
# typecasted Points column as int64 into a new column PointsInt to be able to perform arithmetic sorting
print("The list of teams which came Top 5 in terms of points is:",laliga.nlargest(5,'PointsInt')['Team'].values)
# printed the teams with highest points by using nlargest function
```

```python
# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
# Using the same function, find the team which has maximum and minimum goal difference. (5 points)
laliga['GoalsForInt'] = laliga['GoalsFor'].astype('int64')
laliga['GoalsAgainstInt'] = laliga['GoalsAgainst'].astype('int64')
# typecasting GoalsFor and GoalsAgainst columns into int64
def Goal_diff_count(a,b):
    x = a - b
    return x
# defined the logic for Goals Difference function
laliga['GoalsDiffInt'] = Goal_diff_count(laliga['GoalsForInt'],laliga['GoalsAgainstInt'])
# applied the function to get Goals Difference
laliga.loc[:,['Pos','Team','GoalsDiffInt']]
# prinitng all the teams with their Goal Differences
```

```python
# 5. Create a new column with the name “Winning Percent” and append it to the data set (5 points)
# Percentage of Winning = (GamesWon / GamesPlayed)*100
# If there are any numerical error, replace it with 0%
# Print the top 5 teams which have the highest Winning percentage
laliga['WinningPercent'] = (laliga['GamesWon'].astype('int64')/laliga['GamesPlayed'].astype('int64'))*100
# created a new column with the name “Winning Percent” and appended it to the data set
laliga['WinningPercent'].fillna(0,inplace=True)
# replaced any numerical error with 0%
print("The top 5 teams which have the highest Winning percentage are:",laliga.nlargest(5,'WinningPercent')['Team'].values)
# printed the top 5 teams which have the highest Winning percentage
```

```python
# 6. Group teams based on their “Best position” and print the sum of their points for all positions (7.5 points)
laliga = laliga.apply(pd.to_numeric, errors='ignore')
# converting all the dataframe columns into numeric data type
laliga.groupby(['BestPosition']).sum()['Points']
# grouping the teams based on “Best position” and printed the sum of their points for all positions under Points Column
```
